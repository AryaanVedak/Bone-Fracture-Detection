```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the results file
results_csv = "runs/detect/train_custom/results.csv"  # Update path if necessary
results = pd.read_csv(results_csv)

# Plot training loss
plt.figure(figsize=(10, 6))
plt.plot(results['epoch'], results['train/box_loss'], label="Train Box Loss")
plt.plot(results['epoch'], results['train/cls_loss'], label="Train Class Loss")
plt.plot(results['epoch'], results['train/dfl_loss'], label="Train DFL Loss")
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid()
plt.show()

# Plot validation loss
plt.figure(figsize=(10, 6))
plt.plot(results['epoch'], results['val/box_loss'], label="Validation Box Loss")
plt.plot(results['epoch'], results['val/cls_loss'], label="Validation Class Loss")
plt.plot(results['epoch'], results['val/dfl_loss'], label="Validation DFL Loss")
plt.title("Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid()
plt.show()

# Plot validation metrics
plt.figure(figsize=(10, 6))
plt.plot(results['epoch'], results['metrics/mAP50(B)'], label="mAP@0.5 (B)")
plt.plot(results['epoch'], results['metrics/mAP50-95(B)'], label="mAP@0.5:0.95 (B)")
plt.plot(results['epoch'], results['metrics/precision(B)'], label="Precision (B)")
plt.plot(results['epoch'], results['metrics/recall(B)'], label="Recall (B)")
plt.title("Validation Metrics")
plt.xlabel("Epoch")
plt.ylabel("Metrics")
plt.legend()
plt.grid()
plt.show()
```

```python
# Import necessary libraries for plotting
import os
import matplotlib.pyplot as plt
import random
import cv2
import numpy as np

# Load the class names (replace with the actual class names you have)
class_names = ['Elbow Positive', 'Fingers Positive', 'Forearm Fracture', 'Humerus Fracture', 'Shoulder Fracture', 'Wrist Positive']

# Path to test images directory
test_images_path = r'D:\Project\BoneFractureYolo8\test\images'

# Get a list of all the test image files
test_image_files = os.listdir(test_images_path)

# Select a random sample of test images to display
random_test_images = random.sample(test_image_files, 16)  # Adjust the number as needed

# Set up the plot
fig, axs = plt.subplots(4, 4, figsize=(20, 10))  # Adjust the grid size based on the number of images
axs = axs.flatten()  # Flatten the grid for easier indexing

# Loop over the random images and display them with predictions
for i, image_file in enumerate(random_test_images):
    # Load the image
    image_path = os.path.join(test_images_path, image_file)
    image = cv2.imread(image_path)

    # Run the model's predict method on the image
    results = model.predict(source=image_path, save=False, verbose=False)  # `save=False` avoids saving the output, `verbose=False` suppresses logs
    predictions = results[0]  # Get the first result as it contains the relevant data for this image

    # Draw the bounding boxes on the image
    # `results` is an object that contains predictions, use its method to draw bounding boxes
    for result in results:
        for box in result.boxes:
            x_min, y_min, x_max, y_max = map(int, box.xyxy[0])  # Bounding box coordinates
            class_id = int(box.cls[0])  # Extract class ID
            confidence = box.conf[0].item()  # Extract confidence score

            # Safeguard: Skip invalid class IDs
            if class_id < 0 or class_id >= len(class_names):
                print(f"Invalid class ID {class_id}, skipping this prediction.")
                continue

            # Get the class name
            class_name = class_names[class_id]

            # Draw the bounding box
            cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)

            # Draw the class name and confidence on the image
            label = f"{class_name} {confidence:.2f}"
            label_size, _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 2)
            label_ymin = max(y_min, label_size[1] + 10)
            cv2.rectangle(image, (x_min, label_ymin - label_size[1] - 10), (x_min + label_size[0], label_ymin + 5), (0, 255, 0), -1)
            cv2.putText(image, label, (x_min, label_ymin - 7), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), 2)


    # Convert BGR image to RGB for plotting
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    axs[i].imshow(image_rgb)
    axs[i].axis('off')  # Turn off axes

# Remove empty subplots if fewer images are plotted
for j in range(len(random_test_images), len(axs)):
    axs[j].axis('off')

# Show the plot
plt.tight_layout()
plt.show()
```
